# Splitting the data for training and testing

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

# Model implementation
# building algorithm

```python
from sklearn.linear_model import LogisticRegression

#creating object for class
lr=LogisticRegression()
```

```python
# Model Training

lr.fit(x_train,y_train)
```

```python
#Model Testing

result=lr.predict(x_test)
```

```python
result[:100]
```

```python
y_test
```

# Model Evaluation

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
score=accuracy_score(y_test,result)
print("Accuracy Score For Logistic Model i:",score)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt 
cm=confusion_matrix(y_test,result)
plt.figure(figsize=(8,6),dpi=100)
plt.title("Logistic Regression Matrix")
sns.heatmap(cm,annot=True,fmt='d')
plt.xlabel("Predicted Value")
plt.ylabel("Actual Value")
plt.show()
```

# Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
#model training
dt.fit(x_train,y_train)
#model testing
dt_result=dt.predict(x_test)
dt_result
```

```python
#model evaluation

dt_acc=accuracy_score(y_test,dt_result)
print("Accuracy Score For Decision Tree is:",dt_acc)
```

```python
cm=confusion_matrix(y_test,dt_result)
plt.figure(figsize=(8,6),dpi=100)
plt.title("Decision Tree Classifier Matrix")
sns.heatmap(cm,annot=True,fmt='d')
plt.xlabel("Predicted Value")
plt.ylabel("Actual Value")
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
rf.fit(x_train,y_train)
#model testing
rf_result=rf.predict(x_test)
rf_result
```

```python
rf_acc=accuracy_score(y_test,rf_result)
print("Accuracy Score For Random Forest Classifier is:",rf_acc)
```

```python
cm=confusion_matrix(y_test,rf_result)
plt.figure(figsize=(8,6),dpi=100)
plt.title("Random Tree Classifier Matrix")
sns.heatmap(cm,annot=True,fmt='d')
plt.xlabel("Predicted Value")
plt.ylabel("Actual Value")
plt.show()
```

```python
from sklearn.svm import SVC

sv = SVC()
sv.fit(x_train,y_train)
# model testing
sv_result = sv.predict(x_test)
sv_result
```

```python
sv_acc=accuracy_score(y_test,sv_result)
print("Accuracy Score Support Vector Classifier is:",sv_acc)
```

```python
cm=confusion_matrix(y_test,sv_result)
plt.figure(figsize=(8,6),dpi=100)
plt.title("Support Vector Classifier Matrix")
sns.heatmap(cm,annot=True,fmt='d')
plt.xlabel("Predicted Value")
plt.ylabel("Actual Value")
plt.show()
```
